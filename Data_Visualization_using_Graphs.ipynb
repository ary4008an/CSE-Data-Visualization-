```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print('step 1')
```

```python
import numpy as np
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0, 10, 10)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.plot(x,np.tan(x))
ax.set_title('Trigonometric Funstions')
ax.set_xlabel('Angle')
ax.set_ylabel('Result')
```

```python
import numpy as np
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0, 10, 10)
ax.plot(x,np.sin(x),label='Sin')
ax.plot(x,np.cos(x),label='Cos')
ax.plot(x,np.tan(x),label='Tan')
ax.set_title('Trigonometric Funstions');
ax.set_xlabel('Angle')
ax.set_ylabel('Result')
ax.legend()
plt.show()
```

```python
import numpy as np
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0, 10, 10)
ax.plot(x,np.sin(x),label='Sin',color='black')
ax.plot(x,np.cos(x),label='Cos',color='green')
ax.plot(x,np.tan(x),label='Tan',color='orange')
ax.set_title('Trigonometric Funstions');
ax.set_xlabel('Angle')
ax.set_ylabel('Result')
ax.legend()
plt.show()
```

```python
import numpy as np
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0, 10, 10)
ax.plot(x,np.sin(x),label='Sin',color='black',linestyle='dashed')
ax.plot(x,np.cos(x),label='Cos',color='green',linestyle='dotted')
ax.plot(x,np.tan(x),label='Tan',color='orange',linestyle='dashdot')
ax.set_title('Trigonometric Funstions');
ax.set_xlabel('Angle')
ax.set_ylabel('Result')
ax.legend()
plt.show()
```

```python
import numpy as np
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0, 1, 10)
ax.plot(x,np.sin(x),label='Sin',color='black',linestyle='dashed')
ax.plot(x,np.sin(x),label='Sin(x+1)',color='red',linestyle='dotted')
ax.plot(x,np.cos(x),label='Cos',color='green',linestyle='dotted')
ax.plot(x,np.tan(x),label='Tan',color='orange',linestyle='dashdot')
ax.set_title('Trigonometric Funstions');
ax.set_xlabel('Angle')
ax.set_ylabel('Result')
ax.legend()
plt.show()
```

```python
data={'Apple':10,'Orange':15,'Lemon':5,'Lime':25,'Pomegranate':17,'Banana':9,'Pear':11,'Litchi':69}
names = list(data.keys())
values = list(data.values())
names, values = zip(*sorted(zip(names, values), key=lambda item: item[1], reverse=True))
ax = plt.axes()
ax.barh(names, values)
plt.show()
```

```python
data={'Apple':10,'Orange':15,'Lemon':5,'Lime':25,'Pomegranate':17,'Banana':9,'Pear':11,'Litchi':69}
names = list(data.keys())
values = list(data.values())
names, values = zip(*sorted(zip(names, values), key=lambda item: item[1], reverse=True))
ax = plt.axes()
ax.barh(names, values)
ax.set_title('Fruit Quantities');
ax.set_xlabel('Quantity')
ax.set_ylabel('Fruit')
```

```python
labels=['G1','G2','G3','G4','G5']
men_means=[20,34,30,35,27]
women_means=[25,32,34,20,25]
x=np.arange(len(labels))
width=0.25
fig,ax=plt.subplots()
rects1=ax.bar(x-width/2,men_means,width,label='Men')
rects2=ax.bar(x+width/2,women_means,width,label='Women')
ax.set_title('Scores by group and gender');
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('Scores')
ax.set_xlabel('Gender Categories')
```

```python
labels=['G1','G2','G3','G4','G5']
men_means=[20,34,30,35,27]
women_means=[25,32,34,20,25]
x=np.arange(len(labels))
width=0.25
fig,ax=plt.subplots()
rects1=ax.bar(x-width/2,men_means,width,label='Men')
rects2=ax.bar(x+width/2,women_means,width,label='Women')
ax.set_title('Scores by group and gender');
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('Scores')
ax.set_xlabel('Gender Categories')
ax.bar_label(rects1,padding=3)
ax.bar_label(rects2,padding=3)
ax.set_ylim(0,40)
```
